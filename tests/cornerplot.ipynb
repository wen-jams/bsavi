```python
from bsavi import loaders
import corner
```

```python
params_with_latex = loaders.load_params('../data/planck2018/base_mnu_plikHM_TTTEEE_lowl_lowE_lensing.paramnames')
param_names = list(params_with_latex.keys())
chains = loaders.load_chains('../data/planck2018/*.txt', param_names, params_only=True)
```

```python
chains.keys()
```

```python
model_params = ('omegabh2', 'omegach2')
model_params_with_latex = {key: params_with_latex[key] for key in model_params}
chains = chains.loc[:, model_params]
```

```python
model_params = ('mnu', 'H0*')
model_params_with_latex = {key: params_with_latex[key] for key in model_params}
chains = chains.loc[:, model_params]
```

```python
def format_latex(latex_dict):
    latex_names = list(latex_dict.values())
    formatted_latex = [r'${}$'.format(name) for name in latex_names]
    return formatted_latex
```

```python
formatted_latex = format_latex(model_params_with_latex)
```

```python
# Plot it.
figure = corner.corner(
    chains,
    labels=formatted_latex,
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True,
    title_kwargs={"fontsize": 12},
)
figure.savefig('../images/corner_planck_baryon_vs_cdm.png')
```
